For the BP column (best position), i'll condensate the different positions to make a simpler analysis. And next, i'll drop the column 'Position'.

```python
def clean_bp(x):
    if x in ['ST', 'CF', 'LW', 'RW']:
        return 'ATK'
    elif x in ['CM', 'CAM', 'CDM', 'LM', 'RM']:
        return 'MID'
    elif x in ['LB', 'RB', 'CB']:
        return 'DEF'
    elif x in ['LWB', 'RWB']:
        return 'WING'
    else:
        return 'GK'
```

```python
def height_converting(x):
    if len(x) == 4:
        result = (pd.to_numeric(x[0]) * 30.48) + (pd.to_numeric(x[2]) * 2.54)
    else:
        result = (pd.to_numeric(x[0]) * 30.48) + (pd.to_numeric(x[2:4]) * 2.54)
    return int(result.round(0))
```

```python
# The other method of height converting is resulting in a discrete distribution. I'm trying to see if there is another way
# to make it a continuous distribution, so the it will be closer to a normal distribution and better for our model.

def height_converting2(x):
    if len(x) == 4:
        result = (pd.to_numeric(x[0])) + (pd.to_numeric(x[2]) / 12 )
    else:
        result = (pd.to_numeric(x[0])) + (pd.to_numeric(x[2:4]) / 12)
    return result
```

Columns with a nice looking distribution: age, ova, pot, base stats, passing, dribbling, physical

Theses columns will remain as it is.

Columns weight, attacking, skill, movement, power, mentality, defending, total stats, pace, shooting, defense

These columns are too far away from a normal distribution. They will need to be transformed in order to integrate the linear regression model

Columns with too much outliers: value, wage, release clause, goalkeeping, hits

I don't think the columns value, wage, and release_clause should integrate the model. These informations should not affect the overall rating of a player. The overall rating that should affect it's market value and his wage. So these columns will be dropped.

The column goalkeeping is a special case. It has a weird shape, a long tail, because the majority of the players in this dataset are not goalkeepers, and this brings the distribution to the left. If I apply any kind of filter I will loose all the data for the goalkeepers. So, for this case, I will keep this columns as it is.

Transformations

```python
def log_transf(x):
    return np.log(x)

def power_transf(x):
    return np.sqrt(x)

def power_3_transf(x):
    return np.cbrt(x)

def boxcox_transf(x):
    return pd.Series(stats.boxcox(x)[0])
```

```python
# Columns to transform
# weight, attacking, skill, movement, power, mentality, defending, total_stats, pac, sho, def

columns_to_transform = numericals[['weight', 'attacking', 'skill', 'movement', 'power', 'mentality', 'defending',
              'total_stats', 'pac', 'sho', 'def']]
columns_to_transform.head()
```

```python
for col in columns_to_transform.columns:
    print(col)
    sns.distplot(list(map(log_transf, columns_to_transform[col])))
    plt.show()
```

log_transf worked a litte on these columns: defending, total_stats, sho, def

```python
for col in columns_to_transform.columns:
    print(col)
    sns.distplot(list(map(power_3_transf, columns_to_transform[col])))
    plt.show()
```

neither power_transf or power_3_transf worked on any column

```python
for col in columns_to_transform.columns:
    col_boxcox = boxcox_transf(columns_to_transform[col])
    print(col)
    sns.distplot(col_boxcox)
    plt.show()
```

boxcox_transf worked pretty well on columns: movement, power and pac

List of transformations to make:
- log_transf
    - defending, total_stats, sho, def
- boxcox_transf
    - movement, power, pac

```python
numericals['defending'] = list(map(log_transf, numericals['defending']))
numericals['total_stats'] = list(map(log_transf, numericals['total_stats']))
numericals['sho']= list(map(log_transf, numericals['sho']))
numericals['def'] = list(map(log_transf, numericals['def']))

movement_boxcox = boxcox_transf(numericals['movement'])
numericals['movement'] = movement_boxcox

power_boxcox = boxcox_transf(numericals['power'])
numericals['power'] = power_boxcox

pac_boxcox = boxcox_transf(numericals['pac'])
numericals['pac'] = pac_boxcox
```

### Scalling and Encoding

```python
numerical_features = final_data.select_dtypes(np.number)
numerical_features = numerical_features.drop(['ova'], axis=1) # dropping the target before scalling
numerical_features.head()
```

```python
categorical_features = final_data.select_dtypes(np.object)
categorical_features = categorical_features.drop(['name', 'nationality', 'ir'], axis=1)
# the player name, nationality, and international reputation doesn't affect the target
categorical_features.head()
```

```python
scaler = MinMaxScaler().fit_transform(numerical_features)
scaled = pd.DataFrame(scaler)
scaled
```

The columns w/f, sm, a/w and d/w are already ordinal encoded manually. The only columns to be encoded with OneHotEncoder will be bp (best position), foot and height.

```python
encoder = OneHotEncoder(drop='first').fit_transform(categorical_features[['bp', 'height', 'foot']]).toarray()
encoded = pd.DataFrame(encoder)
encoded
```

```python
# these categorical columns already have ordinal values.
# on the following cells i'll be manually converting the string values and column names to numeric, so the
# linear regression model can make it's calculations

ordinal_encoded = categorical_features[['w/f', 'sm', 'a/w', 'd/w']]
ordinal_encoded.head()
```

```python
ordinal_encoded.dtypes
```

```python
for col in ordinal_encoded.columns:
    ordinal_encoded[col] = ordinal_encoded[col].astype('float')
```

```python
ordinal_encoded.dtypes
```

```python
col_names = []
for i in range(len(ordinal_encoded.columns)):
    col_names.append(i)
col_names
```

```python
ordinal_encoded.columns = col_names
ordinal_encoded.columns
```

```python
features_consolidated = pd.concat([encoded, ordinal_encoded, scaled], axis=1)
```

```python
features_consolidated.head()
```

```python
features_consolidated.shape
```

```python
y = final_data['ova']
y.head()
```

### Linear Regression

```python
x_train, x_test, y_train, y_test = train_test_split(features_consolidated, y, test_size=0.25, random_state=101)
```

```python
print('x_train shape =', x_train.shape)
print('y_train shape =', y_train.shape)
print('x_test shape =', x_test.shape)
print('y_test shape =', y_test.shape)
```

```python
lm = LinearRegression().fit(x_train, y_train)
```

```python
lm.score(x_test, y_test)
```
